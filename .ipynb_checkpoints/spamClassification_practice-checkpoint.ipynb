```python
# Data pre processing
# lower case
#tokenization
#removing of special chracter
# removing of stop words and punction
#stemming
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
import string
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def transform_text(text):
    text = text.lower()  # Convert text to lowercase
    text = nltk.word_tokenize(text)  # Tokenize the text
    y = []  # Initialize an empty list to store processed words

    # Filter out non-alphanumeric tokens
    for i in text:
        if i.isalnum():
            y.append(i)

    # Clear the text variable and reinitialize it with filtered words
    text = y[:]
    y.clear()

    # Remove stopwords and punctuation
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)
    text = y[:]
    y.clear()

    # Perform stemming on remaining words
    for i in text:
        y.append(ps.stem(i))

    return " ".join(y)  # Join the processed words into a single string and return it

```

```python
lemmatizer = WordNetLemmatizer()

def transform_text(text):
    text = text.lower()  # Convert text to lowercase
    text = nltk.word_tokenize(text)  # Tokenize the text
    y = []  # Initialize an empty list to store processed words

    # Filter out non-alphanumeric tokens
    for i in text:
        if i.isalnum():
            y.append(i)

    # Clear the text variable and reinitialize it with filtered words
    text = y[:]
    y.clear()

    # Remove stopwords and punctuation
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)
    text = y[:]
    y.clear()

    # Perform lemmatization on remaining words
    for i in text:
        y.append(lemmatizer.lemmatize(i, pos='v'))  # Lemmatize each word

    return " ".join(y)  # Join the processed words into a single string and return it
```

```python
def get_average_vector(words):
    # Check if words are in the model's vocabulary
    valid_words = [word for word in words if word in model.wv.key_to_index]
    if valid_words:  # If there are valid words, return their average vector
        return sum(model.wv[word] for word in valid_words) / len(valid_words)
    else:
        return np.zeros(model.vector_size)  # Return zero vector if no valid words
```

```python
X = df['transformed_text'].apply(get_average_vector).tolist()
```
